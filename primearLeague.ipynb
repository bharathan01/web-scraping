```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://fbref.com/en/comps/9/schedule/Premier-League-Scores-and-Fixtures"
row_data = requests.get(url)
```

```python
soup = BeautifulSoup(row_data.content,"html")
```

```python
raw_team_title = soup.find_all('th')
print(raw_team_title)
```

```python
team_title = [title.text  for title in raw_team_title[:12] ]
print(team_title)
```

```python
data_frame = pd.DataFrame(columns = team_title)
data_frame
```

```python
table_row = soup.find_all('tr')
table_row
```

```python
for row in table_row[1:]:
    row_table_data = row.find_all('td')
    table_data = [data.text for data in row_table_data[:12]]
    
    df_length = len(data_frame)
    data_frame.loc[df_length] = table_data
    
```
