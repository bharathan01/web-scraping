```python
import pandas as pd
```

```python
df = pd.read_csv('scroes&fixtures.csv')
```

```python

topTeams = ['Arsenal','Brighton','Newcastle Utd','Chelsea',
            'Manchester Utd','Manchester City','Aston Villa','Tottenham','Liverpool','West Ham']
```

```python
filterdData = df[df['Home'].isin(topTeams)]
```

```python
finalData = filterdData[filterdData['Away'].isin(topTeams)]

print(finalData)
```

```python
finalData.to_csv(r'C:\Users\BHARATH\Documents\premeirLeagueDataset\finalDataset.csv', index=False)
```

```python
dropColumn = ['Day', 'Time', 'xG', 'xG.1', 'Attendance', 'Venue', 'Referee']

finalData = finalData.drop(columns = dropColumn)
```

```python
print(finalData)
```

```python
json_data = {
    "FootballScoresAndFixtures": []
}

for index, row in finalData.iterrows():
    match = {
        "date": row['Date'],
        "home": row['Home'],
        "away": row['Away'],
        "score": row['Score']
    }
    json_data["FootballScoresAndFixtures"].append(match)
```

```python
json_data
```

```python
import json

# Define the file path using raw string literal
file_path = r'C:\Users\BHARATH\Documents\premeirLeagueDataset\output.json'

# Write the JSON data to the file
with open(file_path, 'w') as json_file:
    json.dump(json_data, json_file, indent=4)

print("JSON data has been saved to", file_path)
```
